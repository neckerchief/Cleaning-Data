```python
from IPython.display import HTML

# Add this to the first cell of your notebook
display(HTML("""
<style>
  .input_area {
    display: flex;
    flex-direction: column;
    border-left: 5px solid #42a5f5;
    padding-left: 5px;
  }
  
  .input_toggle {
    cursor: pointer;
    border-top: 1px solid #ddd;
    border-bottom: 1px solid #ddd;
    padding: 5px 10px;
    display: flex;
    align-items: center;
    font-family: monospace;
    font-size: 14px;
    user-select: none;
  }
  
  .toggle_arrow {
    display: inline-block;
    margin-right: 5px;
    transition: transform 0.2s;
  }
  
  .toggle_arrow.down {
    transform: rotate(90deg);
  }
  
  .hidden_content {
    display: none;
  }
</style>

<script>
// Run after notebook is fully loaded
document.addEventListener('DOMContentLoaded', function() {
  // Process each code cell
  var cells = document.querySelectorAll('.input');
  
  cells.forEach(function(cell, index) {
    // Create the container
    var cellContent = cell.innerHTML;
    cell.innerHTML = '';
    
    // Create toggle header (like in Kaggle)
    var toggle = document.createElement('div');
    toggle.className = 'input_toggle';
    toggle.innerHTML = '<span class="toggle_arrow">▶</span> <span class="toggle_text">Show code</span>';
    
    // Create content div
    var content = document.createElement('div');
    content.className = 'hidden_content';
    content.innerHTML = cellContent;
    
    // Add to cell
    cell.appendChild(toggle);
    cell.appendChild(content);
    
    // Add toggle functionality
    toggle.addEventListener('click', function() {
      var arrow = this.querySelector('.toggle_arrow');
      var text = this.querySelector('.toggle_text');
      var contentDiv = this.nextElementSibling;
      
      if (contentDiv.classList.contains('hidden_content')) {
        contentDiv.classList.remove('hidden_content');
        arrow.classList.add('down');
        text.textContent = 'Hide code';
      } else {
        contentDiv.classList.add('hidden_content');
        arrow.classList.remove('down');
        text.textContent = 'Show code';
      }
    });
  });
});
</script>
"""))
```
